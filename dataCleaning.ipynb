```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('data/NER/ner_dataset.csv', encoding='unicode_escape')
dataset
```

```python
dataset = dataset.drop(['Sentence #', 'POS'], axis=1)
```

```python
dataset.replace('O', np.nan, inplace=True)
dataset.dropna(subset=['Tag'], inplace=True)

def remove_prefix(x):
    if '-' in x:
        return x.split('-')[1]
    return x

dataset['Tag'] = dataset['Tag'].map(lambda x: remove_prefix(x))
dataset.to_csv('data/NER/ner_onlyTags.csv')
```
